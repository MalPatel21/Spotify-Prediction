FEATURE ENGINEERING
    One hot Encoding the artists_names

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer
```

```python
spotify_sdf=spark.read.parquet('gs://my-bucket-mpat/cleaned/Final_clean_data2.parquet')
```

```python
spotify_sdf.printSchema()
```

```python
indexer = StringIndexer(inputCol="artists_names", outputCol="artists_names_index")
```

```python
indexed_sdf = indexer.fit(spotify_sdf).transform(spotify_sdf)
```

```python
indexed_sdf.select('artists_names','artists_names_index' ).show()
```

```python
encoder = OneHotEncoder(inputCols=["artists_names_index"],
 outputCols=["artists_names_cector"], dropLast=False)
```

```python
encoded_sdf = encoder.fit(indexed_sdf).transform(indexed_sdf)
```

```python
encoded_sdf.select('artists_names','artists_names_index','encoded_sdf' ).show()
```
